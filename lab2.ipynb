**Необходимо вычислить количество бревен и труб на изображениях.**

Решение может быть выполнено на языке python или на языке MATLAB.

Ответ может быть прикреплен к данному заданию в виде файла или может быть отправлена ссылка на репозиторий.

```python
import cv2
import os

# Définir le chemin vers le répertoire contenant les images
# Устанавливаем путь к каталогу, содержащему изображения
repertoire_images = 'C:/Users/moumi/Downloads/task'

# Liste des extensions d'image que vous souhaitez traiter (par exemple, '.jpg', '.png', etc.)
# Список расширений изображений, которые вы хотите обработать (например, «.jpg», «.png» и т. д.)
extensions_images = ['.jpg', '.png']

# Initialiser des compteurs pour les bûches et les canalisations
# Инициализируем счетчики для логов и труб
nombre_total_buches = 0
nombre_total_canalisations = 0

# Parcourir les fichiers dans le répertoire
# Просмотр файлов в каталоге
for fichier in os.listdir(repertoire_images):
    # Vérifier si le fichier a une extension d'image valide
    # Проверьте, имеет ли файл допустимое расширение изображения
    if any(fichier.endswith(extension) for extension in extensions_images):
        # Charger l'image
        image = cv2.imread(os.path.join(repertoire_images, fichier))

        # Convertir l'image en niveaux de gris
        # Преобразуем изображение в оттенки серого
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Appliquer une binarisation pour séparer les objets du fond
        # Применяем бинаризацию для отделения объектов от фона
        _, binary_image = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

        # Trouver les contours des objets
        # Находим очертания объектов
        contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Initialiser des compteurs pour les bûches et les canalisations pour chaque image
        # Инициализируем счетчики логов и каналов для каждого образа
        nombre_de_buches = 0
        nombre_de_canalisations = 0

        # Seuil pour distinguer les bûches des canalisations (à définir en fonction de vos besoins)
        # Порог, позволяющий отличить бревна от труб (определяется в соответствии с вашими потребностями)
        seuil_buche = 4 # пример 4 для журнала, иначе меньше 4 — это как труба
    
        # Parcourir les contours et classer les objets en bûches ou canalisations
        # Просматривайте контуры и классифицируйте объекты по бревнам или трубам
        for contour in contours:
            # j'ai défini mon propre logique pour distinguer les bûches des canalisations.
            # Par exemple, en fonction de la forme, de la taille, de la couleur, etc.
            # Utilise des fonctions OpenCV comme cv2.contourArea() et cv2.approxPolyDP() pour l'analyse des contours.   
            # Я определил свою собственную логику, чтобы отличать журналы от каналов.
            # Например, в зависимости от формы, размера, цвета и т. д.
            # Использует функции OpenCV, такие как cv2.contourArea() и cv2.approxPolyDP(), для анализа контуров.

            # Exemple : Si un contour a une certaine aire, considérez-le comme une bûche
            # Пример: если контур имеет определенную площадь, считайте его бревном.
            if cv2.contourArea(contour) >= seuil_buche:
                nombre_de_buches += 1
            # Sinon, considérez-le comme une canalisation
            else:
                nombre_de_canalisations += 1

        # Afficher le nombre de bûches et de canalisations détectées pour chaque image
        # Показать количество обнаруженных бревен и труб для каждого изображения
        print(f"Изображение: {fichier}")
        print("Количество журналов  :", nombre_de_buches)
        print("Количество труб  :", nombre_de_canalisations)
        print()

        # Ajouter les compteurs de cette image aux compteurs totaux
        # Добавляем счетчики этого изображения к общим счетчикам
        nombre_total_buches += nombre_de_buches
        nombre_total_canalisations += nombre_de_canalisations

# Afficher les totaux pour toutes les images
# Показать итоги для всех изображений
print("Общее количество журналов для всех изображений :", nombre_total_buches)
print("Всего каналов для всех изображений :", nombre_total_canalisations)
```
